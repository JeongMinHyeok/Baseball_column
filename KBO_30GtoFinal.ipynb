```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('30toFinal_Winrate.csv')
```

```python
df.head()
```

```python
x = df['30_Winrate']
y = df['Final_Winrate']
```

```python
b0, b1 = np.polyfit(x, y, 1)
plt.scatter(x, y)
plt.plot(x,b0+b1*x,color='red')
plt.show()
```
